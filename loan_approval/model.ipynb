```python
## importing the packages
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
```

## importing the data

```python
X = pd.read_csv("./X_processed.csv")
y = pd.read_csv("./y_processed.csv")
```

```python
X.head()
```

## split the data

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
```

## building the model

```python
log = LogisticRegression() #instantiating the model instance
log.fit(X_train,y_train) # fitting the model to the data
```

## making predictions

```python
y_pred_test = log.predict(X_test)
```

## evaluating the model

```python
acc = accuracy_score(y_test,y_pred_test)
print(f"Accuracy:{acc}")
```

##  serialization of the model

```python
joblib.dump(log, "loan_approval_linear_model_v1.pkl")
```

## deserialization of the model

```python
final_model = joblib.load("loan_approval_linear_model_v1.pkl")
```

```python
final_model.intercept_, final_model.coef_
```
